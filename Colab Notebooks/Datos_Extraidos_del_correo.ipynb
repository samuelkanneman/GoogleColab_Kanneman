Seleccionar las columnas del cuerpo del archivo correo_filtrado.csv

```python
import pandas as pd

# Cargar el archivo CSV que contiene el cuerpo del mensaje
df = pd.read_csv('/content/correos_filtrados.csv')

# Crear columnas separadas para cada campo
df[['Nombre', 'Apellido']] = df['Cuerpo'].str.extract('Nombre:\s* ([^,]+),Apellido:\s* ([^,]+)')

# Crea la columna 'Email' a partir del contenido de la columna 'cuerpo'
df['Email'] = df['Cuerpo'].str.extract('Email:\s*([^\n]+)')

# Reemplazar la tilde en la letra "e" de "telefono" por una "e" sin tilde
df['Cuerpo'] = df['Cuerpo'].str.replace('tel=C3=A9fono', 'telefono')

# Extraer el número de teléfono del mensaje limpio
df['Numero de telefono'] = df['Cuerpo'].str.extract('Numero de telefono: \s*([^,]+)')


# Seleccionar sólo las columnas deseadas
df = df[['Nombre', 'Apellido', 'Email', 'Numero de telefono']]

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos_mensaje.csv', index=False)
# mostrar mensaje de confirmación
print('El archivo datos_mensaje.csv ha sido creado exitosamente.')
print(df['Nombre'])
```

```python
import pandas as pd
import re

# Leer el archivo CSV
df = pd.read_csv('/content/correos_filtrados.csv')

df.dropna(inplace=True)
df[df['Cuerpo'].isnull()]

# Expresiones regulares para extraer la información deseada
nombre_regex = r'Nombre:\s*(\w+)'
apellido_regex = r'Apellido:\s*(\w+)'
email_regex = r'Email:\s*([\w\.\-]+@[\w\.\-]+)'
telefono_regex = r'Teléfono:\s*([\d+]+)'


# Crear un diccionario para guardar los datos extraídos
datos = {'Nombre': [], 'Apellido': [], 'Email': [], 'Teléfono': []}

# Iterar sobre cada fila de la columna 'Cuerpo'
for texto in df['Cuerpo']:
    # Buscar las coincidencias en el texto
    nombre = re.search(nombre_regex, texto)
    apellido = re.search(apellido_regex, texto)
    email = re.search(email_regex, texto)
    telefono = re.search(telefono_regex, texto)

    # Agregar los datos al diccionario
    datos['Nombre'].append(nombre.group(1) if nombre else '')
    datos['Apellido'].append(apellido.group(1) if apellido else '')
    datos['Email'].append(email.group(1) if email else '')
    datos['Teléfono'].append(telefono.group(1) if telefono else '')

# Convertir el diccionario en un DataFrame
nuevo_df = pd.DataFrame(datos)

# Guardar el DataFrame en un archivo CSV
nuevo_df.to_csv('datos_extraidos.csv', index=False)
```
