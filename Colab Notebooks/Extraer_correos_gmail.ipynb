```python
import imaplib
import email
import csv

# Conectar a tu servidor de correo electrónico
mail = imaplib.IMAP4_SSL('imap.gmail.com')
mail.login('samuelkl.89@gmail.com', 'obxrrnqgajslfiaq')
mail.select('inbox')

# Buscar correos electrónicos de un remitente específico
result, data = mail.search(None, 'FROM', 'samuelkl.89', 'SUBJECT', 'presupuesto')

# Crear una lista vacía para guardar los datos extraídos
data_list = []

# Iterar sobre cada correo electrónico y extraer los datos necesarios
for email_id in data[0].split():
    _, data = mail.fetch(email_id, '(RFC822)')
    message = email.message_from_bytes(data[0][1])
    
    # Verificar si el mensaje tiene varias partes
    if message.is_multipart():
        # Recorrer la lista de partes y extraer el contenido de cada parte
        partes = []
        for parte in message.get_payload():
            if parte.get_content_type() == 'text/plain':
                partes.append(parte.get_payload())
        # Unir el contenido de las partes
        mensaje = '\n'.join(partes)
    else:
        # Extraer el contenido del mensaje
        mensaje = message.get_payload()
    
    # Limpiar el contenido del mensaje
    mensaje_limpio = mensaje.replace('\r\n\r\n', '\r\n')
    mensaje_limpio = mensaje_limpio.replace('\r\n', ',')
    
    data_dict = {
        'Correo': message['From'],
        'Mensaje': mensaje_limpio
    }
    data_list.append(data_dict)

# Crear un DataFrame a partir de la lista de datos
df = pd.DataFrame(data_list)

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos_correo.csv', index=False)

# Cerrar la conexión con el servidor de correo electrónico
mail.close()
mail.logout()
```

De datos_correo.csv extraer en limpio el mensaje en columnas

```python
import pandas as pd

# Cargar el archivo CSV que contiene el cuerpo del mensaje
df = pd.read_csv('/content/datos_correo.csv')

# Crear columnas separadas para cada campo
df[['Nombre', 'Apellidos']] = df['Mensaje'].str.extract('Nombre: ([^,]+),Apellidos: ([^,]+)')
df['numero de DNI'] = df['Mensaje'].str.extract('numero de DNI: ([^,]+)')

# Reemplazar la tilde en la letra "e" de "telefono" por una "e" sin tilde
df['Mensaje'] = df['Mensaje'].str.replace('tel=C3=A9fono', 'telefono')

# Extraer el número de teléfono del mensaje limpio
df['Numero de telefono'] = df['Mensaje'].str.extract('Numero de telefono: ([^,]+)')

# Seleccionar sólo las columnas deseadas
df = df[['Nombre', 'Apellidos', 'numero de DNI', 'Numero de telefono']]

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos_mensaje.csv', index=False)
```
