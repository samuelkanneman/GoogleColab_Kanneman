```python
nombres = ['Pedro', 'Juan', 'Diego']
cromosomas = ['00000101010101010101', '00101010101101110111', '00100010010000001001']
```

```python
# obtener la muestra del usuario
muestra = input('Ingrese secuencia: ')

# inicializar variables para el cromosoma mÃ¡s parecido
mejor_cromosoma = ''
mejor_nombre = ''
mejor_porcentaje = 0

# comparar la muestra con cada cromosoma en la lista
for i in range(len(cromosomas)):
    cromosoma = cromosomas[i]
    nombre = nombres[i]
    genes_iguales = 0
    for j in range(len(muestra)):
        if muestra[j] == cromosoma[j]:
            genes_iguales += 1
    porcentaje = genes_iguales / len(muestra) * 100
    if porcentaje > mejor_porcentaje:
        mejor_cromosoma = cromosoma
        mejor_nombre = nombre
        mejor_porcentaje = porcentaje

# mostrar el resultado
print(f'El culpable es {mejor_nombre} con un parentezco de {mejor_porcentaje:.0f}%')
```
