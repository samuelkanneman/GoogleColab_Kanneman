```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://bonistas.com/"
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
articulos = soup.find_all("h2", class_="titulo-articulo")
for articulo in articulos:
    print(articulo.text)
```

```python
# Encontrar el div específico utilizando un selector o atributo
div_especifico = soup.find("div", class_="table-responsive table-sm")  # Reemplaza "nombre_del_div" con el ID del div que deseas encontrar

# Verificar si se encontró el div y mostrar su contenido
if div_especifico:
    print('Se encontro!')
else:
    print("No se encontró el div específico")
```

```python
# Crear una lista vacía para almacenar los datos de la tabla
datos_tabla = []

# Extraer los datos de la tabla del div
filas = div_especifico.find_all("tr")
for fila in filas:
    celdas = fila.find_all("td")
    fila_datos = [celda.text.strip() for celda in celdas]
    datos_tabla.append(fila_datos)

# Crear un DataFrame de pandas con los datos de la tabla
df = pd.DataFrame(datos_tabla)
```

```python
# Guardar el DataFrame como un archivo CSV
nombre_archivo = "tabla_datos.csv"  # Nombre del archivo CSV que deseas guardar
df.to_csv(nombre_archivo, index=False)
```
