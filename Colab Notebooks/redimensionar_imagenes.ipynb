```python
from PIL import Image
import os

# directorio de las imágenes originales
input_dir = '/content/viejas'

# directorio para guardar las imágenes reducidas
output_dir = '/content/nuevas'
print('Exito')
```

reducir ingresando las dimensiones originales

```python


# tamaño deseado para las imágenes reducidas
new_size = (int(800*0.9), int(600*0.9))

# número de imágenes por paquete
batch_size = 8

# obtener la lista de archivos en el directorio de entrada
file_list = os.listdir(input_dir)

# procesar las imágenes en paquetes de tamaño batch_size
for i in range(0, len(file_list), batch_size):
    # obtener el lote de imágenes
    batch_files = file_list[i:i+batch_size]

    # procesar cada imagen en el lote
    for filename in batch_files:
        # abrir la imagen original
        img = Image.open(os.path.join(input_dir, filename))

        # redimensionar la imagen
        img = img.resize(new_size, Image.ANTIALIAS)

        # guardar la imagen redimensionada en el directorio de salida
        output_path = os.path.join(output_dir, filename)
        img.save(output_path, optimize=True, quality=85)
```

reducir sin conocer las dimensiones originales

```python
import os

# directorio de entrada
input_dir = '/content/viejas'

# directorio de salida
output_dir = '/content/nuevas'

# asegurarse de que el directorio de salida existe
os.makedirs(output_dir, exist_ok=True)
```

```python
import os
from PIL import Image

# tamaño deseado para las imágenes reducidas
scale_factor = 0.9

# número de imágenes por paquete
batch_size = 8

# obtener la lista de archivos en el directorio de entrada
file_list = os.listdir(input_dir)

# procesar las imágenes en paquetes de tamaño batch_size
for i in range(0, len(file_list), batch_size):
    # obtener el lote de imágenes
    batch_files = file_list[i:i+batch_size]

    # procesar cada imagen en el lote
    for filename in batch_files:
        # abrir la imagen original
        img = Image.open(os.path.join(input_dir, filename))

        # obtener el tamaño original de la imagen
        original_size = img.size

        # calcular el nuevo tamaño de la imagen reducida
        new_size = (int(original_size[0] * scale_factor), int(original_size[1] * scale_factor))

        # redimensionar la imagen
        img = img.resize(new_size, Image.ANTIALIAS)

        # guardar la imagen redimensionada en el directorio de salida
        output_path = os.path.join(output_dir, filename)
        img.save(output_path, optimize=True, quality=85)
```
