```python
import speech_recognition as sr
import docx

# Inicializar un objeto recognizer
r = sr.Recognizer()

# Abrir el archivo de audio y reconocer el texto
for i in range(25):
    with sr.AudioFile(f'seccion_{i}.wav') as source:
        audio = r.record(source)
    text_section = r.recognize_google(audio, language='es-ES')
    
    # Guardar la transcripci√≥n en un archivo de Microsoft Word
    document = docx.Document()
    document.add_paragraph(text_section)
    document.save(f"transcripcion_{i}.docx")
```
