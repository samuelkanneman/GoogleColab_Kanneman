Eliminar duplicados de la columna email

```python
# Leer el archivo CSV
df = pd.read_csv('/content/correos2.CSV')

# Eliminar los duplicados para la columna Email
df.drop_duplicates(subset=['De: (dirección)'], keep='first', inplace=True)

# Guardar el archivo sin duplicados
df.to_csv('archivo_sin_duplicados2.csv', index=False)


# Guardar el archivo sin duplicados en formato Excel
df.to_excel('archivo_sin_duplicados2.xlsx', index=False)
print('Exito')
```

correos duplicados

```python
# Leer el archivo .xlsx
df = pd.read_excel('/content/archivo_limpiado.xlsx')

# Verificar si hay correos electrónicos duplicados
duplicados = df.duplicated(subset=['Email'], keep=False)
filas_duplicadas = df[duplicados].sort_values(by=['Email'])

# Mostrar las filas duplicadas en la columna de correo electrónico
if len(filas_duplicadas) > 0:
    print('Se encontraron correos electrónicos duplicados:')
    print(filas_duplicadas)
else:
    print('No se encontraron correos electrónicos duplicados.')
```

limpiar duplicados

```python
# Leer el archivo .xlsx
df = pd.read_excel('/content/archivo_limpiado.xlsx')

# Eliminar las filas duplicadas en la columna de correo electrónico
df_sin_duplicados = df.drop_duplicates(subset=['Email'], keep='first')

# Guardar el archivo limpio en un nuevo archivo .xlsx
df_sin_duplicados.to_excel('archivo_sin_duplicados.xlsx', index=False)

print('Archivo guardado sin filas duplicadas.')
```
