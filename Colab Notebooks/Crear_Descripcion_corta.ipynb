```python
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('/content/descripcion_corta.csv')

# Iterar sobre cada fila del DataFrame
for i, row in df.iterrows():
    # Verificar si la descripción contiene "Código:"
    if 'Código:' in row['Descripcion Corta']:
        # Encontrar el índice de la cadena "Código:"
        codigo_idx = row['Descripcion Corta'].index('Código:')
        # Insertar el código correspondiente después de la cadena "Código:"
        nuevo_codigo = row['Descripcion Corta'][:codigo_idx+7] + ' ' + str(codigos_array[i]) + row['Descripcion Corta'][codigo_idx+7:]
        # Actualizar el valor en la columna "Nuevo Código"
        df.at[i, 'Descripcion Corta'] = nuevo_codigo

# Guardar el DataFrame actualizado en un nuevo archivo CSV
df.to_csv('archivo_actualizado.csv', index=False)
```
