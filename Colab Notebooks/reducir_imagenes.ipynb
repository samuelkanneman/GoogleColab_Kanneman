```python
from PIL import Image

# abrir la imagen
img = Image.open("imagen.jpeg")

# reducir el tama√±o de la imagen a la mitad
new_width = int(img.size[0] / 2)
new_height = int(img.size[1] / 2)
img = img.resize((new_width, new_height), Image.ANTIALIAS)

# guardar la imagen reducida
img.save("imagen_reducida.jpeg", optimize=True, quality=85)
```
