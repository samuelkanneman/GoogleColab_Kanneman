```python
# Crear una lista de palabras comunes que deseas eliminar
stop_words = ["y", "de", "la", 'un', 'una', 'o','del','que', 'otro']

# Función para eliminar las palabras comunes
def eliminar_stop_words(texto):
    palabras = texto.split()
    palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]
    return " ".join(palabras_filtradas)

# Eliminar las palabras comunes del texto
texto_filtrado = eliminar_stop_words(texto)
```

```python
# Configurar WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_filtrado)

# Mostrar la nube de palabras
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
