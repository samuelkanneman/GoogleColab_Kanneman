```python
from pydub import AudioSegment
import math

def divide_audio(audio, num_secciones):
    duracion = len(audio)
    seccion_duracion = math.floor(duracion / num_secciones)

    for i in range(num_secciones):
        inicio_seccion = i * seccion_duracion
        fin_seccion = (i + 1) * seccion_duracion
        seccion = audio[inicio_seccion:fin_seccion]
        nombre_archivo = f'seccion_{i}.wav'
        seccion.export(nombre_archivo, format='wav')

# Cargar el archivo de audio
audio = AudioSegment.from_wav("Grabacion.wav")

# Solicitar la cantidad de secciones
num_secciones = int(input("Ingrese el n√∫mero de secciones en las que desea dividir el audio: "))

# Dividir el audio
divide_audio(audio, num_secciones)
```
